```python
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import numpy as np
```

```python
huge_df = pd.read_csv("data/census_health_ff_2011_2016.csv")
huge_df = huge_df.loc[huge_df["County"] != "Alpine"]
huge_df = huge_df.loc[huge_df["County"] != "Mono"]
huge_df.head()
list(huge_df.columns)
```

```python
plt.scatter(huge_df["Fast-food restaurants/10,000 pop, 2011"], huge_df["Obesity 2011 %"])
plt.title("Fast Food restaurants/10,000 pop vs Obesity Rate 2011")
plt.xlabel("Fast Food restaurants/10,000 pop")
plt.ylabel("Obesity Rate")
plt.grid(True)

(vc_slope, vc_int, vc_r, vc_p, vc_std_err) = stats.linregress(huge_df["Fast-food restaurants/10,000 pop, 2011"], huge_df["Obesity 2011 %"])
linear_reg = vc_slope * huge_df["Fast-food restaurants/10,000 pop, 2011"] + vc_int
plt.plot(huge_df["Fast-food restaurants/10,000 pop, 2011"],linear_reg,"--")
plt.savefig("charts/food_obesity_2011.png")
plt.show()
```
