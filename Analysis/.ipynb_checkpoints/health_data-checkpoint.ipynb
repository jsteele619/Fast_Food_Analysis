```python
import pandas as pd
import csv
```

```python

obesity_2011 = pd.read_csv("data/obesity2011.csv", header=2)
obesity_2016 = pd.read_csv("data/obesity2016.csv", header=2)
new_diabetes_2011 = pd.read_csv("data/new_diabetes2011.csv", header=2)
new_diabetes_2016 = pd.read_csv("data/new_diabetes2016.csv", header=2)
diagnosed_diabetes_2011 = pd.read_csv("data/diagnosed_diabetes2011.csv", header=2)
diagnosed_diabetes_2016 = pd.read_csv("data/diagnosed_diabetes2016.csv", header=2)

big_df = pd.read_csv("data/new_fast_food_combined_data.csv")
```

```python
big_df["Obesity 2011 %"] = obesity_2011["Percentage"]
big_df["Obesity 2016 %"] = obesity_2016["Percentage"]
big_df["New Diabetes per 1k : 2011"] = new_diabetes_2011["Rate per 1000"]
big_df["New Diabetes per 1k : 2016"] = new_diabetes_2016["Rate per 1000"]
big_df["Diagnosed Diabetes 2011 %"] = diagnosed_diabetes_2011["Percentage"]
big_df["Diagnosed Diabetes 2016 %"] = diagnosed_diabetes_2016["Percentage"]
```

```python
big_df.to_csv("data/big_df.csv", index=False)
big_df.head()
```
